Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#created a dataframe to hold the dataset
ccdf=pd.read_csv('creditcard.csv')
ccdf.head()#showed the first 5 rows
ccdf.tail()#showed the last 5 rows
```

```python
#getting information about the dataset
ccdf.info()

```

```python
#check for missing values
ccdf.isnull().sum()
```

```python
#distribution of legitimate and fraud transactions
ccdf['Class'].value_counts()
```

The dataset we chose is highly imbalanced, the class indicates a fraudulent transaction for value 1.0 and a legitimate transaction for value 0.0

```python
legit=ccdf[ccdf.Class==0]
fraud=ccdf[ccdf.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare both the types of transcations
ccdf.groupby('Class').mean()
```

```python
#Build a new dataset having similar distribution of normal transactions and fraudulent transactions
legit_sample=legit.sample(n=492)#since number of fraud=492

```

```python
#join both dataframes to get a new dataset
newdf=pd.concat([legit_sample,fraud],axis=0)
#display first 5 rows
newdf.head()
```

```python
newdf.tail()
```

```python
newdf.value_counts()
```

```python
newdf.groupby('Class').mean()
```

```python
X=newdf.drop(columns='Class',axis=1)
Y=newdf['Class']
print(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
X_test_prediction=model.predict(X_test)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on training dataset:",training_data_accuracy)
print("Accuracy on test dataset:",test_data_accuracy)
```
